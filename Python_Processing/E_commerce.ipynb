```python
import pandas as pd
# Define file path
file_path = 'ecommercedataset20232024 (4).csv'
try:
    # Load raw data
    df = pd.read_csv(file_path)
    print("Data loaded successfully. Starting data prep...")
    # 1. Data Cleaning
    # Drop duplicated order_id
    initial_rows = len(df)
    df = df.drop_duplicates(subset=['order_id'])
    if initial_rows - len(df) > 0:
        print(f"Dropped {initial_rows - len(df)} duplicate rows.")
    # Parse order_date to datetime format
    df['order_date'] = pd.to_datetime(df['order_date'])
    # Handle null values in return_reason
    df['return_reason'] = df['return_reason'].fillna('No Return')
    # Strip whitespaces on categorical columns
    cols_to_strip = ['channel', 'device', 'payment_method', 'product_category']
    for col in cols_to_strip:
        if df[col].dtype == 'object':
            df[col] = df[col].str.strip()
    # 2. Feature Engineering
    # Categorize discount levels
    df['discount_tier'] = pd.cut(
        df['discount_pct'],
        bins=[-1, 5, 15, 100],
        labels=['Low', 'Medium', 'High']
    )
    # Flag profit status
    df['profit_status'] = df['gross_profit'].apply(lambda x: 'Profit' if x > 0 else 'Loss')
    # Categorize session engagement
    df['session_engagement'] = pd.cut(
        df['session_duration_min'],
        bins=[0, 3, 10, 1000],
        labels=['Short', 'Medium', 'Long']
    )
    # Categorize shipping duration
    df['shipping_category'] = pd.cut(
        df['shipping_days'],
        bins=[0, 3, 7, 100],
        labels=['Fast', 'Standard', 'Slow']
    )
    # Flag loyalty status based on repeat purchases
    df['loyalty_status'] = df['repeat_purchases'].apply(lambda x: 'Loyal' if x >= 3 else 'Regular')
    # 3. Export to Excel
    output_file = 'Master_Data_Ecommerce_Cleaned.xlsx'
    with pd.ExcelWriter(output_file, engine='openpyxl', datetime_format='YYYY-MM-DD') as writer:
        df.to_excel(writer, sheet_name='Master_Data', index=False)
    print(f"Data preparation complete. Saved as {output_file}")
except Exception as e:
    print(f"Error executing script: {e}")
```
